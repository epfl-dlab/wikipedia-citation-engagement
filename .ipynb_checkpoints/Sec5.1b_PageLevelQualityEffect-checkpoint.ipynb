```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
import seaborn as sns
```

```python
features = pd.read_pickle("extended_page_features.pkl")
```

```python
len(features)
```

```python
features.head(3)
```

```python
quality_map = {
    'Stub': 0,
    'Start': 0.2,
    'C': 0.4,
    'B': 0.6,
    'GA': 0.8,
    'FA': 1.0
}
```

```python
features_list = []

for index, row in features.dropna().iterrows():
    quality = row['quality']
    if quality is not None and row["pageviews"]>0:
        top_quality = sorted([(name, value) for name, value in quality.items()], key=lambda r: -r[1])[0]
        features_list.append({"page_ct": row["page_ct"], 
                              "title": row["title"], 
                              "pageviews": row["pageviews"], 
                              "quality": quality_map[top_quality[0]]})
```

```python
quality_df = pd.DataFrame(features_list)
quality_df.head(3)
```

```python
len(quality_df)
```

```python
import math
quality_df["popularity_class"] = pd.qcut(quality_df.pageviews, 10, labels=False)

quality_df.head()
```

```python
popularities = quality_df.popularity_class.unique()
```

```python
from scipy.stats import binned_statistic
import numpy as np
import builtins
import math

def get_stats(bin_samples, stat_type="mean"):
    if len(bin_samples)<1:
        return 
    if stat_type=="mean":
        return np.mean(bin_samples)
    means_list = []
    for i in range(0, 100):
        samples = np.random.choice(bin_samples, size=len(bin_samples))
        means_list.append(np.mean(samples))
    means_list = sorted(means_list)
    if stat_type=="left":
        left_index = int(0.05*len(means_list))
        return means_list[left_index]
    elif stat_type=="right":
        right_index = builtins.min(len(means_list)-1, math.ceil(0.95*len(means_list)))
        return means_list[right_index]

def get_binned_stats(x, y, bins):
    binned = binned_statistic(x, y, 
                     bins=bins, 
                     statistic=lambda r: get_stats(r, "mean"))
    binned_low = binned_statistic(x, y, 
                     bins=bins, 
                     statistic=lambda r: get_stats(r, "left"))
    binned_up = binned_statistic(x, y,
                     bins=bins, 
                     statistic=lambda r: get_stats(r, "right"))
   
    return pd.Series(binned.bin_edges).rolling(2, 2).mean().dropna(),\
            binned.statistic, binned_low.statistic, binned_up.statistic
```

```python
import math
bins = [-0.1, 0.1, 0.3, 0.5, 0.7, 0.9, 1.1]
row_count = math.ceil(len(popularities)/2)

fig, axs = plt.subplots(row_count, 2, figsize=(8,10))
popularities = sorted(popularities)
for idx in range(0, len(popularities)):
    p = popularities[idx]
    pages=quality_df[quality_df.popularity_class==p]
    bin_x, mean_values, lower, upper = get_binned_stats(pages.quality, pages.page_ct, bins)
    axs[idx//2][idx%2].plot(bin_x, mean_values, linestyle='--', marker=".")
    axs[idx//2][idx%2].fill_between(bin_x, lower,upper, color = 'gray',alpha = 0.3)
    axs[idx//2][idx%2].set_title('Popularity group {} - [{:.0f}, {:.0f}]'.format(p,
                                                                   quality_df.pageviews.quantile(idx/10),
                                                                   quality_df.pageviews.quantile((1+idx)/10)
                                                                  ))
    axs[idx//2][idx%2].set(ylabel='Page CT Rate')
    axs[idx//2][idx%2].set_ylim([0,0.015])

fig.tight_layout()
```

```python
quality_df[quality_df.popularity_class==8].groupby('quality').count()['pageviews']\
    .plot(kind="bar", title="Pages distribution in popularity group 8")
plt.xlabel("Quality")
plt.ylabel("Number of pages")
```

```python
ax = sns.boxplot(x="quality", y="page_ct", data=quality_df[quality_df.popularity_class==8], 
                 showfliers=False, showmeans=True).set_title('Popularity group 8')
plt.xlabel("Quality")
plt.ylabel("Page CT rate")
```

```python
ax = sns.boxplot(x="quality", y="pageviews", data=quality_df[quality_df.popularity_class==8], 
                 showfliers=False, showmeans=True).set_title('Popularity group 8')
plt.xlabel("Quality")
plt.ylabel("Page views")
```
